##### The problem and its significance. 

The bicycle industry has boomed within the last 2 years of a global pandemic. With bicycles, parts and accessories being scarces sales and repairs have been affected along with unusual wait times, delivery delays and more.

The following project simulates a 'typical' day at a bicycle shop and the services they provide pre-pandemic. 

Note: this does exclude phone calls and emails from customers and focusing only on in person interactions:

* Customer Arrival
* Meet with person working the floor: Depending on the type of service needed the customer will be going into one of three different service areas:
    * Random inquiries: using air pump, questions not related to the business, browsing, etc
    * Sales: looking to buy a bicycle, parts or accessories
        * Special orders: if items are not available some customers may decided to place an order with us (can take a week or more depending on product availability for the order to arrive)
    * Repairs: within this category there's different types of services
        * Short repair: flat fixes, minor adjustments (these usually take anywhere between 10 - 45 minutes)
        * Long repairs: tune-ups, bike assemblies, part replacements, major adjustments (these usually take between 1 hour - 3 business days)
* Storage: if repairs don't get picked up during the time frame specified then they are moved to storage after 30 days (they tend to stay there until the bicycles get thrown out after 1 year of not being claimed)
* Normal Exit

```python
# Required libraries
import simpy
import numpy as np
import pandas as pd
import random
from modsim import *
import statistics
```

```python
# Setting seed and iterations
seed = 42
simulation_time = 8 # operational hours
```

```python
np.random.seed(seed)

report = pd.DataFrame(columns = ['Customer ID', 'Service', 'Arrival Time', 'Service Start Time', 'Service Stop Time'])
```

```python
# Setting customer arrival every 15 minutes from 2 - 5 customers
def customer_arrival():
    return np.random.exponential(15 / np.random.randint(2, 6))
    
# Setting time it takes for a customer to meet with floor person   
def meet_floor_person(self, customer):
    # min of 2 minutes and max of 10 minutes
    yield self.env.timeout(random.randint(2, 11)) 
        
# Setting time it takes for customer to be in repairs
def do_repairs(self, customer):
    # min of 10 minutes and max of 3 days in minutes
    yield self.env.timeout(random.randint(10, 4321)) 
    
# Setting time it takes for customer to be in sales
def make_sales(self, customer):
    # min of 15 minutes and max of 45 minutes
    yield self.env.timeout(random.randint(15, 46)) 
    
# Setting time it takes for customer to be in random_inq
def random_inq(self, customer):
    # min of 2 minutes and max of 10 minutes
    yield self.env.timeout(random.randint(2, 11))    
    
# Setting time it takes for customer's bike to be in storage
def bike_storage(self, customer):
    # min of 7 days and max of 365 days in minutes
    yield self.env.timeout(random.randint(10080, 525601))  
        
# Setting time it takes for customer to exit
def exit(self, customer):
    # min of 10 minutes and max of 1 hour in minutes
    yield self.env.timeout(random.randint(10, 61))
        
```

```python
# Defining customer's arrival time
def bike_shop(env, repairs, sales, random_inq):
    global current_hour
    i = 0
    
    while True:
        Service = np.random.choice(['Repairs', 'Sales', 'Random Inquiries'], p = [0.85, 0.1, 0.05])
        i += 1
        yield env.timeout(customer_arrival())
        env.process(customer(env, i, repairs, sales, random_inq, Service))
```

```python
def customer(env, customer, repairs, sales, random_inq, Service):
    global report

    if Service == 'Repairs':
        req = repairs
    else:
        req = sales
        
    with repairs.request() as request:
        arrival_time = env.now
        yield env.timeout(10)
        yield request
        service_start_time = env.now
        
        if Service == 'Repairs':
            yield env.timeout(repairs_service())
        else:
            yield env.timeout(sales_service())
        service_stop_time = env.now
        
    row = pd.DataFrame([['Customer_{}'.format(customer), Service, arrival_time, service_start_time, service_stop_time]],
                          columns = ['Customer Id', 'Service', 'Arrival Time', 'Service Start Time', 'Service Stop Time'])
    report = report.append(row, ignore_index = True, sort = False)
        
    #Calculations for wait time
    report['Service Time (Mins)'] =  (report['Service Stop Time'] - report['Service Start Time'])
    report['Wait Time (Mins)'] =  ( report['Service Start Time'] - report['Arrival Time'] )
    
```

```python
#Running it in a model
def model(nruns, nrepairs, nsales, nrandom_inq): #pass in no of runs for simulation, capacity teller, capacity exchange
    #initialise
    Repairs = [] 
    Sales = []
    Random_inq = [] 
    avg_wait_time = []
    repairs_wait_time = []
    sales_wait_time = []
    random_inq_wait_time = []
    
    
    global report
    for j in range(nruns):
        report = report[0:0]
        env = simpy.Environment()
        env.initial_time = 0
        
        # resource
        repairs = simpy.Resource(env, capacity = nrepairs)
        sales = simpy.Resource(env, capacity = nsales)
        random_inq = simpy.Resource(env, capacity = nrandom_inq)
        env.process(bike_shop(env, repairs, sales, random_inq))
        env.run(until = 8 * 60) #run for 8 hours
        
        
        Repairs = report[report.Service.isin(['Repairs'])] #separate to calculate mean later
        Sales = report[report.Service.isin(['Sales'])]
        Random_inq = report[report.Service.isin(['Random Inquiries'])]
    
        avg_wait = report['Wait Time (Mins)'].mean() #Calculate average time
        avg_wait_repairs = Repairs['Wait Time (Mins)'].mean()
        avg_wait_sales = Sales['Wait Time (Mins)'].mean()
        avg_wait_random_inq = Random_inq['Wait Time (Mins)'].mean()

    
        avg_wait_time.append(avg_wait)
        repairs_wait_time.append(avg_wait_repairs)
        sales_wait_time.append(avg_wait_sales)
        random_inq_wait_time.append(avg_wait_random_inq)
        
    #Save in a dataframe
    new_dict = {"avg_wait": avg_wait_time, "repairs_wait": repairs_wait_time, "sales_wait": sales_wait_time, "random_inq_wait": random_inq_wait_time}
    
    global new_df
    new_df = pd.DataFrame(new_dict)
    fig=new_df.boxplot(grid=False, widths=0.5)
    return new_df
    return fig
```

```python
np.random.seed(68)
model(30,5,1)
```
